# Reliability diagrams across every dataset

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.calibration import calibration_curve
```

```python
df_german = pd.read_csv('German Credit-y_results.csv')
df_card_default = pd.read_csv('Credit Card Default-y_results.csv')
df_term_deposit = pd.read_csv('Bank Term Deposit-y_results.csv')
```

```python
def plot_reliability_diagram(dataset_name, model_methods):
    n_models = len(model_methods)
    fig, axes = plt.subplots(1, n_models, figsize=(6 * n_models, 5), sharex=True, sharey=True)
    # If only one model, wrap axes in a list
    if n_models == 1:
        axes = [axes]
    else:
        fig.suptitle(f"{dataset_name} - Reliability Diagrams", y=0.98)

    for ax, (model_name, methods) in zip(axes, model_methods.items()):
        for (method_name, y_true, y_proba) in methods:
            prob_true, prob_pred = calibration_curve(y_true, y_proba, n_bins=10)
            ax.plot(prob_pred, prob_true, marker='o', label=method_name)

        # Perfect calibration line
        ax.plot([0, 1], [0, 1], '--', color='gray', label='Perfect Calibration')
        ax.set_title(f"{model_name}")
        ax.set_xlabel("Predicted probability")
        ax.set_ylabel("Observed frequency")
        ax.legend()
        ax.grid()

    plt.tight_layout()
    plt.show()
```

```python
df = df_german
plot_reliability_diagram(
    'German Credit',
    {
        'Logistic Regression': [
          ('Uncalibrated', df['y_test'], df['Logistic Regression y_test_proba']),
          ('Platt Scaling', df['y_test'], df['Logistic Regression y_platt_proba']),
          ('Isotonic Regression', df['y_test'], df['Logistic Regression y_isotonic_proba']),
          ('Beta Calibration', df['y_test'], df['Logistic Regression y_beta_proba']),
      ],
      'Random Forest': [
          ('Uncalibrated', df['y_test'], df['Random Forest y_test_proba']),
          ('Platt Scaling', df['y_test'], df['Random Forest y_platt_proba']),
          ('Isotonic Regression', df['y_test'], df['Random Forest y_isotonic_proba']),
          ('Beta Calibration', df['y_test'], df['Random Forest y_beta_proba']),
      ],
      'XGBoost': [
          ('Uncalibrated', df['y_test'], df['XGBoost y_test_proba']),
          ('Platt Scaling', df['y_test'], df['XGBoost y_platt_proba']),
          ('Isotonic Regression', df['y_test'], df['XGBoost y_isotonic_proba']),
          ('Beta Calibration', df['y_test'], df['XGBoost y_beta_proba']),
      ],
    }
)
```
